```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Dual-Line Plot with Markers and Labels
months=['Jan','Feb','Mar','Apr','May','Jun']
revenue=[10000,7500,12000,12500,11000,8500]
profit=[5000,3750,6000,6250,5500,4250]

plt.plot(months, revenue, marker='o', linestyle='-', color='r', label='Revenue')
plt.plot(months, profit, marker='x', linestyle='--', color='b', label='Profit')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Revenue and Profit over 6 months')
plt.legend()
plt.show()
```

```python
#Vertical Bar Chart with Annotations
grades=['1','2','3','4','5']
books_read=[10,12,12,13,15]
colors = ['blue', 'red', 'green', 'yellow', 'purple']
bars = plt.bar(grades, books_read, color=colors)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{height}', ha='center', va='bottom')
plt.xlabel('Grade')
plt.ylabel('Books Read')
plt.title('Books Read by Students in Different Grades')
plt.show()
```

```python
#Customized Pie Chart
brands = ['Samsung', 'Apple', 'Xiaomi', 'Oppo', 'Honor']
market_share = [40, 25, 20, 10, 5]
colors = ['blue', 'red', 'orange', 'green', 'purple']
explode = [0.1 if i == max(market_share) else 0 for i in market_share]
plt.pie(market_share, labels=brands, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Market Share of Smartphone Brands')
plt.show()
```

```python
#Subplots with Shared Y-axis
days = [ 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
hours_studied = [2, 4, 4, 3, 5, 2, 4]
hours_sleep = [8, 9, 7, 7, 6, 8, 8]
fig,(study, sleep) = plt.subplots(1, 2, sharey=True, figsize=(12, 6))
study.plot(days, hours_studied, marker='o', linestyle='-', color='red', label='Hours Studied')
study.set_title('Hours Studied Over a Week')
study.set_xlabel('Day of the Week')
study.set_ylabel('Hours')
sleep.plot(days, hours_sleep, marker='s', linestyle='--', color='blue', label='Hours of Sleep')
sleep.set_title('Hours of Sleep Over a Week')
sleep.set_xlabel('Day of the Week')
plt.tight_layout()
plt.show()
```

```python
#Stacked Bar Chart with Labels
tasks = ['Extraction', 'Cleaning', 'Analysis', 'Visualization']
team_1 = [7, 6, 8, 9]
team_2 = [8, 5, 7, 8]
team_3 = [6, 8, 8, 9]
colors = ['blue', 'green', 'red']
bars_A = plt.bar(tasks, team_1, color=colors[0], label='Team 1')
bars_B = plt.bar(tasks, team_2, bottom=team_1, color=colors[1], label='Team 2')
bars_C = plt.bar(tasks, team_3, bottom=[i + j for i, j in zip(team_1, team_2)], color=colors[2], label='Team 3')
for bar in bars_A:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height/2.0, f'{height}', ha='center', va='center', color='black')
for bar, a in zip(bars_B, team_1):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, a + height/2.0, f'{height}', ha='center', va='center', color='black')
for bar, a, b in zip(bars_C, team_1, team_2):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, a + b + height/2.0, f'{height}', ha='center', va='center', color='black')
plt.xlabel('Tasks')
plt.ylabel('Performance')
plt.title('Performance of Teams in Different Tasks')
plt.legend()
plt.show()
```

```python
#Complex Pie Chart with Nested Categories
categories = ['Housing', 'Food', 'Transportation']
expenses = [40, 30, 30]
subcategories = ['Rent', 'Utilities', 'Groceries', 'Dining Out', 'Gas', 'Public Transport']
sub_expenses = [25, 15, 20, 10, 15, 15]
inner_colors = ['red', 'green', 'blue']
outer_colors = ['orange', 'pink', 'limegreen', 'darkgreen', 'skyblue', 'deepskyblue']
fig, ax = plt.subplots(figsize=(8, 8))
explode = [0.1, 0.1, 0.1]
ax.pie(expenses, labels=categories, colors=inner_colors, radius=1, wedgeprops=dict(width=0.3, edgecolor='w'), autopct='%1.1f%%', explode=explode)
ax.pie(sub_expenses, labels=subcategories, colors=outer_colors, radius=0.7, wedgeprops=dict(width=0.3, edgecolor='w'), autopct='%1.1f%%')
plt.title('Family Expenses Breakdown')
plt.show()
```
